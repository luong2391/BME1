### Tạo vòng lặp để thực hiện tạo ra 50 phương trình tuyến tính với Y lần lượt là các loại thuốc 1-51 và X là 50 loại còn lại

```python
dt.head()
```

```python
for i in range(51):
    X = dt.drop("{}".format(i+1), axis = 1) #chọn cột
    Y = dt.iloc[:,i]
    lm = LinearRegression()
    lm.fit(X, Y)
    
    
    print("Equation {}".format(i+1))            #sai số bình phương trung bình
    print(lm.coef_)              #hệ số của các biến độc lập
    print(lm.intercept_) 
```
