```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import sklearn
import statsmodels.api as sm
import array as arr
import seaborn as sns
sns.set_style("whitegrid")
sns.set_context("poster")

# special matplotlib argument for improved plots
from matplotlib import rcParams
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

## Tên các thuốc được đánh dấu từ 1 - 51, dữ liệu số là tổng số lượng thuốc bán được theo quý (2019/1/1-2022/9)

```python
dt = pd.read_csv('data.csv')
#dt.set_index("Nước cất ống nhựa", inplace=True)
dt
```

### Tạo phương trình tuyến tính

```python
def linear(X, Y):
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.4, random_state = 2)
    print(X_train.shape)
    print(X_test.shape)
    print(Y_train.shape)
    print(Y_test.shape)

    lm = LinearRegression()
    lm.fit(X_train, Y_train)
    Y_pred = lm.predict(X_test)

    plt.scatter(Y_test, Y_pred)
    plt.xlabel("Prices: $Y_i$")
    plt.ylabel("Predicted prices: $\hat{Y}_i$")
    plt.title("Prices vs Predicted prices: $Y_i$ vs $\hat{Y}_i$")
    mse = sklearn.metrics.mean_squared_error(Y_test, Y_pred)
    print(mse)                   #sai số bình phương trung bình
    print(lm.coef_)              #hệ số của các biến độc lập
    print(lm.intercept_)         #hệ số tự do
    
    return linear(X, Y)
    
```

### Tạo vòng lặp để thực hiện tạo ra 50 phương trình tuyến tính với Y lần lượt là các loại thuốc 1-51 và X là 50 loại còn lại

```python
temp = [1]
for i in range(1, 50):
    temp[0]= temp[0]+1
    if i == temp[0]:
        X = dt.drop(temp,axis=1 )
        Y = dt[i]
        X.set_index("2", inplace = True)
    else: 
        print('none')
       # X = dt.drop(columns = [i])
        #Y = dt['i']
       # X.set_index("1", inplace = True)
return linear(X,Y)
# mình chưa biết làm vòng lặp này
```
